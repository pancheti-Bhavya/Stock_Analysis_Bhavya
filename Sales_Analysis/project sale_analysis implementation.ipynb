```python
import pandas as pd
aapl_stock=pd.read_csv("aapl_stock_prices.csv")
goog_stock=pd.read_csv("goog_stock_prices.csv")
msft_stock=pd.read_csv("msft_stock_prices.csv")
```

```python
#replacing missing value(close_price) with median close price in aapl_stock
aapl_stock['close_price'] = aapl_stock['close_price'].fillna(aapl_stock['close_price'].median())
aapl_stock
```

```python
#replacing missing value(Volume) with 0 in aapl_stock
aapl_stock['volume'] = aapl_stock['volume'].fillna(0)
aapl_stock
```

```python
#replacing missing value(close_price) with median close price in goog_stock
goog_stock['close_price'] = goog_stock['close_price'].fillna(goog_stock['close_price'].median())
goog_stock
```

```python
#replacing missing value(Volume) with 0 in goog_stock
goog_stock['volume'] = goog_stock['volume'].fillna(0)
goog_stock
```

```python
#replacing missing value(close_price) with median close price in msft_stock
msft_stock['close_price'] = msft_stock['close_price'].fillna(msft_stock['close_price'].median())
msft_stock
```

```python
#replacing missing value(Volume) with 0 in msft_stock
msft_stock['volume'] = msft_stock['volume'].fillna(0)
msft_stock
```

```python
#deleting the duplicates values in aapl_stock
aapl_stock = aapl_stock.drop_duplicates()
aapl_stock
```

```python
##deleting the duplicates values in goog_stock
goog_stock = goog_stock.drop_duplicates()
goog_stock
```

```python
#deleting the duplicates values in msft_stock
msft_stock = msft_stock.drop_duplicates()
aapl_stock
```

```python
# Apply Price Sanity Checks
aapl_stock["high_price"] = aapl_stock[
    ["high_price", "open_price", "close_price"]
].max(axis=1)

aapl_stock["low_price"] = aapl_stock[
    ["low_price", "open_price", "close_price"]
].min(axis=1)
```

```python
# Apply Price Sanity Checks
goog_stock["high_price"] = goog_stock[
    ["high_price", "open_price", "close_price"]
].max(axis=1)

goog_stock["low_price"] = goog_stock[
    ["low_price", "open_price", "close_price"]
].min(axis=1)
```

```python
# Apply Price Sanity Checks
msft_stock["high_price"] = msft_stock[
    ["high_price", "open_price", "close_price"]
].max(axis=1)

msft_stock["low_price"] = msft_stock[
    ["low_price", "open_price", "close_price"]
].min(axis=1)
```

```python
#Calculate daily return
aapl_stock["daily_return"] = (
    (aapl_stock["close_price"] - aapl_stock["open_price"])
    / aapl_stock["open_price"]
)
aapl_stock
```

```python
#Calculate daily return
goog_stock["daily_return"] = (
    (goog_stock["close_price"] - goog_stock["open_price"])
    / goog_stock["open_price"]
)
goog_stock
```

```python
#Calculate daily return
msft_stock["daily_return"] = (
    (msft_stock["close_price"] - msft_stock["open_price"])
    / msft_stock["open_price"]
)
msft_stock
```

```python
#Treands Classification
aapl_stock["trend"] = aapl_stock["daily_return"].apply( lambda x: "UP" if x > 0 else "DOWN" if x < 0 else "NO_CHANGE" )
aapl_stock
```

```python
#Treands Classification
goog_stock["trend"] = goog_stock["daily_return"].apply( lambda x: "UP" if x > 0 else "DOWN" if x < 0 else "NO_CHANGE" )
goog_stock
```

```python
#Treands Classification
msft_stock["trend"] = msft_stock["daily_return"].apply( lambda x: "UP" if x > 0 else "DOWN" if x < 0 else "NO_CHANGE" )
msft_stock
```

```python
#Calculate volatility standard deviation of daily returns
aapl_volatility = aapl_stock["daily_return"].std()
goog_volatility = goog_stock["daily_return"].std()
msft_volatility = msft_stock["daily_return"].std()
print("AAPL Volatility:", aapl_volatility)
print("GOOG Volatility:", goog_volatility)
print("MSFT Volatility:", msft_volatility)
```

```python
#Identify highest volatility stock
volatility_dict = {
    "AAPL": aapl_volatility,
    "GOOG": goog_volatility,
    "MSFT": msft_volatility
}
highest_volatility_stock = max(volatility_dict, key=volatility_dict.get)
highest_volatility_stock
```

```python
aapl_corr = aapl_stock["volume"].corr(aapl_stock["daily_return"])
goog_corr = goog_stock["volume"].corr(goog_stock["daily_return"])
msft_corr = msft_stock["volume"].corr(msft_stock["daily_return"])

print("AAPL Volume vs Price Correlation:", aapl_corr)
print("GOOG Volume vs Price Correlation:", goog_corr)
print("MSFT Volume vs Price Correlation:", msft_corr)
```

```python
#Moving averages were calculated to smooth price trends and identify short-term and long-term movements in stock prices
#I used 7-day(blue) and 30-day(orange) moving averages to analyze short-term and long-term stock trends
aapl_stock = aapl_stock.sort_values("date")
aapl_stock["MA_7"] = aapl_stock["close_price"].rolling(window=7).mean()
aapl_stock["MA_30"] = aapl_stock["close_price"].rolling(window=30).mean()
plt.plot(aapl_stock["date"], aapl_stock["MA_7"], label="7-Day MA")
plt.plot(aapl_stock["date"], aapl_stock["MA_30"], label="30-Day MA")
plt.title("AAPL Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```
